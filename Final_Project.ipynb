```python
# Import the core libraries needed for data loading, analysis, NLP setup, and GPU-enabled PyTorch
import os
import re
import sys
import time
import random

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import transformers
import accelerate

from urllib.request import urlretrieve
from urllib.parse import quote

print("Python executable:", sys.executable)
print(torch.__version__)
print("CUDA available:", torch.cuda.is_available())
print("Torch CUDA version:", torch.version.cuda)
print("GPU:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")
```

```python
# Set the download folder to the current working directory so the project files save in the active notebook location
download_dir = os.getcwd()
print("Current working directory:", download_dir)
```

```python
# Setting the Seed for Replication of Results
SEED = 7
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)
```

```python
# Create a writable download folder and define the cloud file locations used in the baseline notebook
download_dir = os.getcwd()
os.makedirs(download_dir, exist_ok=True)

base_url = "https://iu-abjm-cloud-prod.s3.us-east-2.amazonaws.com"
files = {
    "games_description": "Games Description.csv",
    "games_ranking": "Games Ranking.csv",
    "steam_game_reviews": "Steam Game Reviews.csv",
}

# Download each CSV file and store the local file paths for later loading
local_paths = {}

for name, filename in files.items():
    url = f"{base_url}/{quote(filename)}"
    local_path = os.path.join(download_dir, filename.replace(" ", "_"))
    print(f"Downloading: {url}")
    urlretrieve(url, local_path)
    local_paths[name] = local_path

print("\nDownloaded files:")
for name, path in local_paths.items():
    print(f"{name}: {path}")
```
